```python
# get rows having date +- a week of December 12th
part1 = data[(data.Stock_Creation_Date >= '2017-12-5') & 
             (data.Stock_Creation_Date <= '2017-12-19')]

# drop rows having such dates
sample2 = data.drop(data[(data['Stock_Creation_Date'] >= '2018-08-8') 
                        & (data['Stock_Creation_Date'] <= '2018-08-22')].index)
```

```python
# get rows having date +- a week of August 15th

part2 = data[(data.Stock_Creation_Date >= '2018-01-19') & (data.Stock_Creation_Date <= '2018-02-2')]
part2 = part2.Stock_Creation_Date.value_counts().sort_index()

# drop rows having such dates for sample2
sample2 = data.drop(data[(data.Stock_Creation_Date >= '2018-01-19') 
                         & (data.Stock_Creation_Date <= '2018-02-2')].index)
```

```python
# get rows having date +- a week of December 26th
part3 = data[(data.Stock_Creation_Date >= '2018-02-7') & (data.Stock_Creation_Date <= '2018-02-21')]
part3 = part3.Stock_Creation_Date.value_counts().sort_index()

# drop rows having such dates for sample2
sample2 = data.drop(data[(data.Stock_Creation_Date >= '2018-02-7') 
                         & (data.Stock_Creation_Date <= '2018-02-21')].index)
```

```python
# get rows having date +- a week of December 14th
part4 = data[(data.Stock_Creation_Date >= '2018-08-8') & (data.Stock_Creation_Date <= '2018-08-22')]
part4 = part4.Stock_Creation_Date.value_counts().sort_index()

# drop rows having such dates for sample2
sample2 = data.drop(data[(data.Stock_Creation_Date >= '2018-08-8') 
                         & (data.Stock_Creation_Date <= '2018-08-22')].index)
```

```python

from scipy import stats
part34 = data[(data.Stock_Creation_Date >= '2018-03-01') & (data.Stock_Creation_Date <= '2018-04-30')]
part56 = data[(data.Stock_Creation_Date >= '2018-05-01') & (data.Stock_Creation_Date <= '2018-06-30')]
part34=part34.Stock_Creation_Date.value_counts().sort_index()
part56=part56.Stock_Creation_Date.value_counts().sort_index()
stats.ttest_ind(a=part56,b=part34, axis=0,nan_policy='propagate', equal_var=False, alternative='greater')
```
